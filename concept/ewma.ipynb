# exponentially  weighted moving average

>It’s a type of moving average that gives more importance (weight) to recent data points and less importance to older ones.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = [10, 13, 12, 15, 14]
df = pd.DataFrame({'value': data})

# Compute EWMA
df['EWMA'] = df['value'].ewm(alpha=0.5, adjust=False).mean()

# Display
print(df)

# Plot
plt.plot(df['value'], label='Original Data')
plt.plot(df['EWMA'], label='EWMA (α=0.5)', linewidth=3)
plt.legend()
plt.show()
```
